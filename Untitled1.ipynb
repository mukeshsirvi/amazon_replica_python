```python
import csv
class amazon:
    
    def __init__(self, user, seller):
        self.user = user
        self.seller = seller
        self.menu()
        
    def menu(self):
        print('''----Select your choice----
                1. Register
                2. Login
                 ''' )
        choice = int(input("Enter your choice : "))
        if choice == 1:
            self.register()
        elif choice == 2:
            self.login()
        else:
            print("Wrong Choice selected")
            
    def register(self):
        with open("seller_data.txt","a") as f:
            write = csv.writer(f)
            
            user_name = input("Enter your name : ")
            password = input("Enter your password : ")
            email = input("Enter your Email Id : ")
            phone = int(input("Enter your phone number : "))
            add = input("Enter your address: ")
            write.writerow([user_name, password, email, phone,add])
            print("Registraion Sucessful")
            self.menu()
            
    def login(self):
    
        not_log_in = True
        while (not_log_in == True):
            with open("seller_data.txt","r") as f:
                user_name = input("Enter your User name :")
                password = input("Enter your password: ")
                reader = csv.reader(f)
                for row in reader:
                    for col in row:
                        if (col == user_name and row[1] == password):
                            not_log_in = False
                        else:
                            break
                    
                if (not_log_in == True):
                    print("Wrong username or  password! Try Again")
                else:
                    print("Access granted")
                    self.product()
                    
    def product(self):
        print('''Select your choice :
        1. Create product
        2. Update product
        3. view product
        4. Delete product
        5. Exit ''')
        choice = int(input("Select your choice "))
        if choice == 1:
            self.create_product()
        elif choice == 2:
            self.update_product()
        elif choice == 3:
            self.view_product()
        elif choice == 4:
            self.delete_product()
        elif choice == 5:
            self.exit()
        else:
            print("You entered wrong choice")
            
        
        
    def create_product(self):
        print("----Enter details regarding product---- ")
        with open("product_details.txt","a") as f:
            write = csv.writer(f)
        
            product_name = input("Enter Product Name : ")
            product_id = input("Product Id :")
            manufacture_name = input("Enter Manufaturer Name :")
            price = int(input("Enter price : "))
            quantity = int(input("Enter Quantity :"))
            
            write.writerow([product_name,product_id,manufacture_name,price,quantity])
            print("Your product is created")
            self.product()
            
            
    def update_product(self):
        with open("product_details.txt","r") as f:
            reader = csv.reader(f)
            product_id = input("Enter the product id you want to update :")
            for row in reader:
                for col in row:
                    if col[2] == product_id:
                        self.create_product()
                    else:
                        print("Product id is incorrect")
                        
        
        self.product()
                        
    def view_product(self):
        with open("product_details.txt","r") as f:
            reader = csv.reader(f)
            for row in reader:
                for col in row:
                    print(col,end=" ")
                    
        self.product()
    
    def delete_product(self):
        pass
    
    def exit(self):
        print("App Exit ")
        
                

        
    
```
